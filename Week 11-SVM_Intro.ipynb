```python
X = np.array([ [1.0, 0.0], [5.0, 0.0], [2.0, 10.0], [6.0, -1.0]])
y = np.array( [1, -1, 1, -1])

random.seed(100)
N = 1000
XX = random.randn(N, 2) + [12, 10]
XXX = random.randn(N, 2) + [5, 6]
X = np.concatenate((X, XX, XXX))

yy = [-1 ] * N
yyy = [1 ] * N
y = np.append(y, yy)
y = np.append(y, yyy)

xmin=-10; xmax=15
ymin=-20; ymax=50
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.scatter(X[:,0], X[:,1], c=y);

model = svm.SVC(C = 1, gamma=0.000001, kernel='linear')
time_start = time.time()
model.fit(X, y)
print('train time =', time.time()-time_start)

print(model.intercept_, model.coef_)
print(model.intercept_.shape, model.coef_.shape)
a=model.coef_[0, 0]
b=model.coef_[0, 1]
c= model.intercept_[0]
if( np.abs(a) < np.abs(b)) :
    Xseperate = random.uniform(xmin, xmax,1000 )
    Yseperate = -1.0 *a * Xseperate / b - c / b
else:
    Yseperate = random.uniform(ymin, ymax,1000 )
    Xseperate = -1.0 *b * Yseperate / a - c / a

# print(Xseperate, ' ', Yseperate)
plt.plot(Xseperate, Yseperate,'g')

pred = model.predict(X)
result= pred-y
print(result.sum())
print(pred)
print(model.score(X, y))
```
